## **READ DATA**

**We download the outputs we got from running kape in the szechuan evidence.**

```python
url = 'https://drive.google.com/uc?id=1j6AyNWyrgkwGxEPcttYwbuIf7EzQmhto'
of = 'kape.tgz'
gdown.download(url, of, quiet=False)
gdown.extractall("kape.tgz", to="kape/")
```

**Let's read the kape outputs with the xread() function from ds4n6_lib. It will generate the .pkl file we will use with the dataframes and store the collection of dataframes in d4.out**

```python
%%time

d4.debug = 0

# Read raw data
xread(tool="kape", rootpath=home_path[0]+"/kape", folder_parsing_mode="single_host_with_categories", path_prefix="G__szechuan_kape_artifacts", pluginisdfname=True, notebook_file=thisnb)
```

**We store the dataframes in a more friendly variable: kpdfs.**

```python
kpdfs = d4.out
```
