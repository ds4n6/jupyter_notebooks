**We download the outputs we got from running autoruns in the szechuan evidence.**

```python
url = 'https://drive.google.com/uc?id=1oddrNtlIfhWSm773yKZ-uZwPgKrOUnUq'
of = 'autoruns.tgz'
gdown.download(url, of, quiet=False)
gdown.extractall("autoruns.tgz", to="autoruns/")
```

**Let's read the autoruns outputs with the xread() function from ds4n6_lib. It will generate the .pkl file we will use with the dataframes and store the collection of dataframes in d4.out**

```python
%%time

d4.debug = 0

# Read raw data
xread(tool="autoruns", rootpath=home_path[0]+"/autoruns", path_prefix="dc01", usepickle=False, encoding='utf-16', notebook_file=thisnb)
```
